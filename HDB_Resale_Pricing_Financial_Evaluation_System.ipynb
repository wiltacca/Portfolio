Using relevant functions from the Numpy package, you are to develop a simple Python program to compute the correlation coefficient between the HDB Resale Price Index (RPI) and Singapore’s population from 1990 to 2019.

Note: You are required to extract all the datapoints only from the fourth quarters spanning 1990 to 2019 from the provided list of HDB RPI data. These datapoints are to be used as yearly HDB RPI datapoints necessary for computation of the correlation with the yearly population levels provided. You will also need to scale down the population values provided, dividing them by 1,000,000 (round off results to 2 decimals).

```python
# Insert your code for part (A) here.

# Use these provided lists of data for HDB's quarterly RPI and Singapore's yearly population spanning 1990 to 2019
rpi = [24.3, 24.4, 25.0, 24.7, 24.9, 25.5, 25.2, 25.1, 25.8, 27.4, 28.0, 28.6, 30.2, 39.6, 47.7, 49.0, 50.4, 51.7, 54.6, 54.8, 57.3, 63.9, 67.7, 73.7, 80.5, 90.9, 96.4, 99.0, 98.6, 97.5, 93.6, 88.0, 81.7, 78.3, 74.9, 72.3, 71.2, 72.2, 78.1, 79.8, 80.3, 79.3, 77.6, 75.8, 73.2, 72.1, 70.6, 69.6, 69.1, 69.2, 69.9, 69.9, 71.0, 72.5, 74.3, 75.1, 75.3, 76.2, 76.3, 77.1, 77.2, 73.5, 73.2, 73.5, 73.6, 74.3, 74.2, 74.9, 75.8, 78.1, 83.2, 88.0, 91.3, 95.4, 99.4, 100.8, 100.0, 101.4, 105.0, 109.0, 112.1, 116.6, 121.3, 124.4, 126.4, 130.4, 135.4, 137.7, 138.5, 140.3, 143.1, 146.7, 148.6, 149.4, 148.1, 145.8, 143.5, 141.5, 139.1, 137.0, 135.6, 135.0, 134.6, 134.8, 134.7, 134.7, 134.7, 134.6, 133.9, 133.7, 132.8, 132.6, 131.6, 131.7, 131.6, 131.4, 131.0, 130.8, 130.9, 131.5]
population = [3047132, 3135083, 3230698, 3313471, 3419048, 3524506, 3670704, 3796038, 3927213, 3958723, 4027887, 4138012, 4175950, 4114826, 4166664, 4265762, 4401365, 4588599, 4839396, 4987573, 5076732, 5183688, 5312437, 5399162, 5469724, 5535002, 5607283, 5612253, 5638676, 5703569]

import numpy as np

# Use for division 
million = 1_000_000

# Extract rpi datapoints only from fourth quarters, dividing them by 1,000,000 and round off to 2 decimals
np_rpi = np.array(rpi[3:121:4])
np_population = np.round_(np.array(population) / million, decimals = 2)

# numpy.column_stack() function is used to stack 1-D arrays as columns into a 2-D array
np_col_stack = np.column_stack((np_rpi, np_population))
print("2-D Array of Yearly RPI and Population Levels Since 1990 \n", np_col_stack)

print()

print("The shape of this array is", np_col_stack.shape)

# Compute the correlation coefficient and round off to 3 decimals
conrelation = (np.corrcoef(np_col_stack[:,0], np_col_stack[:,1]))
print(f"The correlation between the yearly RPI and yearly population is", np.round_(conrelation[0 ,1], decimals = 3))
```
